```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
features = pd.read_csv('../Datasets/uci_har/features.txt',header=None,sep=' ',names=('ID','Activity'))
labels = pd.read_csv('../Datasets/uci_har/activity_labels.txt',header=None,sep=' ',names=('ID','Sensor'))

X_train = pd.read_table('../Datasets/uci_har/train/X_train.txt',header=None,sep='\s+')
y_train = pd.read_table('../Datasets/uci_har/train/y_train.txt',header=None,sep='\s+')

X_test = pd.read_table('../Datasets/uci_har/test/X_test.txt',header=None,sep='\s+')
y_test = pd.read_table('../Datasets/uci_har/test/y_test.txt',header=None,sep='\s+')

train_sub = pd.read_table('../Datasets/uci_har/train/subject_train.txt',header=None,names=['SubjectID'])
test_sub = pd.read_table('../Datasets/uci_har/test/subject_test.txt',header=None,names=['SubjectID'])
```

```python
X_train.head()
```

```python
X_train.columns = features.iloc[:,1]
X_test.columns = features.iloc[:,1]
```

```python
y_train.columns = ['Activity']
y_test.columns = ['Activity']
```

```python
X_train['SubjectID'] = train_sub
X_test['SubjectID'] = test_sub
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.metrics import classification_report,confusion_matrix,mean_squared_error
```

```python
from sklearn.model_selection import KFold,StratifiedKFold 
```

```python
from sklearn.model_selection import RepeatedKFold
```

```python
def rmsle(y, y_pred):
    return np.sqrt(mean_squared_error(y, y_pred))
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
y_pred = dtree.predict(X_test)
```

```python
print(rmsle(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
random_forest = RandomForestClassifier(n_estimators=6)
```

```python
kfold = KFold().split(X_train,y_train)
```

```python
for k, train in enumerate(kfold):
    random_forest.fit(X_train[train], y_train[train])
```
